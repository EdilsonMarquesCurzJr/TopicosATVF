## 2) Análise univariada

### Idade


### Idade

A distribuição da idade dos respondentes mostra um forte predomínio de jovens adultos entre 25 e 35 anos, fato que indica que o mercado de dados ainda é composto, em sua maioria, por profissionais em início ou meio de carreira.


```python
# HISTOGRAMA
plt.figure()
sns.histplot(df['idade'], bins=20, kde=True)
plt.title('Distribuição de Idade dos Respondentes', fontsize=16)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()
```

```python
# BOXPLOT
plt.figure(figsize=(8, 8))
sns.boxplot(data=df['idade'],palette='colorblind')

plt.title("Boxplot da Idade")
plt.tight_layout()
plt.show()
```

```python
print(df['idade'].describe())
print(f"Moda: {df['idade'].mode()[0]}")
print(f"Assimetria (Skewness): {scipy.stats.skew(df['idade']):.2f}")
print(f"Curtose: {scipy.stats.kurtosis(df['idade']):.2f}")
print("\n")

# Teste Anderson-Darling
ad_result = scipy.stats.anderson(df['idade'], dist='norm')
print(f"Teste Anderson-Darling")
print(f"Estatística: {ad_result.statistic:.4f}")
for sl, cv in zip(ad_result.significance_level, ad_result.critical_values):
    print(f"Nível de significância: {sl}%, Valor crítico: {cv}")

if ad_result.statistic < ad_result.critical_values[2]:  # índice 2 = nível 5%
    print("Conclusão: Provavelmente distribuição NORMAL")
else:
    print("Conclusão: Provavelmente NÃO é distribuição NORMAL")

print("\n-----------------------------\n")

# Teste D’Agostino’s K² (normaltest)
k2_stat, k2_pvalue = scipy.stats.normaltest(df['idade'])
print("Teste D’Agostino’s K²")
print(f"Estatística: {k2_stat:.4f}, p-valor: {k2_pvalue:.4f}")

if k2_pvalue > 0.05:
    print("Conclusão: Provavelmente distribuição NORMAL")
else:
    print("Conclusão: Provavelmente NÃO é distribuição NORMAL")
```

> A faixa de idade das pessoas que responderam a pesquisa se encontra entre 18-69 anos, porém possuindo uma quantidade maior de "jovens", como demonstrado pelo 3º quartil, que mostra que 75% dos indivíduos tem até 36 anos

> A média de idade é de 32 anos, enquanto que a mediana é de 31, indicando uma leve assimetria positiva, confirmada tanto pela skewness, quanto pelo histograma e boxplot, que mostram que a pequena quantidade de idades mais altas interfere na distribuição de idades. De tal modo, os testes estatísticos também confirmam a probabilidade maior da distribuição não ser de cunho normal.

> A moda de 29 anos também indica uma alta quantidade de participantes mais jovens

### Faixa salarial

```python
try:
    salary_order = sorted(df['faixa_salarial'].dropna().unique(), key=lambda x: int(x.split(' a ')[0].replace('R$ ', '').replace('.', '')))
except:
    salary_order = df['faixa_salarial'].value_counts().index

sns.countplot(y=df['faixa_salarial'], order=salary_order, palette='viridis')
plt.title('Distribuição por Faixa Salarial', fontsize=16)
plt.xlabel('Contagem', fontsize=12)
plt.ylabel('Faixa Salarial (R$)', fontsize=12)
plt.show()
```

```python
df['faixa_salarial'].value_counts()
```

> A distribuição sugere que o público da pesquisa é majoritariamente de profissionais com salários médios a altos, o que é compatível com perfis de profissionais de tecnologia e ciência de dados, que costumam ter remunerações acima da média nacional.

> A presença de uma quantidade significativa nas faixas mais altas pode indicar uma parcela experiente ou em posições de destaque.

> A existência de pessoas nas faixas mais baixas mostra diversidade na amostra, possivelmente incluindo iniciantes ou pessoas em transição de carreira.

> A faixa “Não informado” deve ser considerada ao analisar a distribuição para evitar viés.



### Localidade


```python
df_coluna = df['uf_onde_mora'].copy()

ax = sns.countplot(y=df_coluna, order=df_coluna.value_counts().index, palette='colorblind')

total = len(df_coluna)

for p in ax.patches:
    width = p.get_width()  # valor absoluto da barra
    percentual = width / total * 100
    ax.text(width + total*0.01,  # um espacinho depois da barra
            p.get_y() + p.get_height() / 2,  # centralizado verticalmente na barra
            f'{int(width)} ({percentual:.1f}%)',
            ha='left', va='center', fontsize=9)

plt.tight_layout()
plt.subplots_adjust(left=0.25)

plt.title('Distribuição de participantes por UF de residência')
plt.show()
```

> São Paulo (SP) lidera disparado, com 2.064 participantes, quase 40% do total, refletindo seu peso econômico e tecnológico no país, já que é o maior polo de tecnologia e negócios do Brasil.

> Outros estados com números relativamente altos são:
- Minas Gerais (MG) com 580 participantes,
- Paraná (PR) com 415,
- Rio de Janeiro (RJ) com 397,
- Rio Grande do Sul (RS) com 369,
mostrando que o Sudeste e Sul concentram a maior parte dos respondentes.

> O Distrito Federal (DF) também aparece com uma quantidade relevante: 186 participantes.

> Há uma parcela de 139 pessoas morando fora do Brasil, o que indica que a pesquisa alcançou um público internacional ou expatriados.

> Estados do Nordeste e Norte aparecem com números menores, como Ceará (CE) 128, Pernambuco (PE) 120, Bahia (BA) 99, Amazonas (AM) 23, e outros com menos de 100 participantes.

> Estados com menor participação, como Amapá (AP) com 4 e Tocantins (TO) com 6, sugerem menor representatividade regional.

```python
df_coluna = df['nivel_de_ensino']
sns.countplot(data=df_coluna, order=df_coluna.value_counts().index, palette='colorblind')
```

### Área de formação


```python
df_coluna = df['area_de_formacao'].copy()

ax = sns.countplot(y=df_coluna, order=df_coluna.value_counts().index, palette='colorblind')

total = len(df_coluna)

for p in ax.patches:
    width = p.get_width()  # valor absoluto da barra
    percentual = width / total * 100
    ax.text(width + total*0.01,  # um espacinho depois da barra
            p.get_y() + p.get_height() / 2,  # centralizado verticalmente na barra
            f'{int(width)} ({percentual:.1f}%)',
            ha='left', va='center', fontsize=9)

plt.tight_layout()
plt.subplots_adjust(left=0.25)

plt.title('Distribuição de participantes por Área de Formação')
plt.show()
```

> A maioria dos participantes tem formação em áreas de computação, o que sugere uma forte afinidade e facilidade de transição para o mercado de dados, já que as competências técnicas se alinham.

> As áreas de engenharia, administração e negócios também possuem um alto número de participantes, indicando uma compatibilidade com o campo de dados, que beneficia-se de raciocínio analítico e visão estratégica.

> A formação em matemática e estatística, embora importante, representa uma parcela menor de participantes, contrariando a percepção comum de que a área de dados é dominada por matemáticos.

> Áreas como ciências sociais, biológicas, saúde, química e física têm pouca representatividade na amostra, mostrando que a diversidade de formação ainda é limitada na área.

> A participação de pessoas sem formação superior é muito baixa, reforçando a alta qualificação do grupo estudado.
